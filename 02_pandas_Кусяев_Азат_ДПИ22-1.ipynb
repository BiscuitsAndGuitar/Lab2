### Объединение таблиц `pd.DataFrame`

5.1 При помощи объединения таблиц, создайте `DataFrame`, состоящий из четырех столбцов: `id`, `name`, `user_id`, `rating`. Рецепты, на которые не оставлен ни один отзыв, должны отсутствовать в полученной таблице. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и попытавшись найти строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
#Создаю версию таблицы, где присутствуют только те рецепты, которые имеют отзывы
recipes_with_reviews = reviews[reviews['review'].notna()]
recipes_with_reviews
```

```python
id_and_name = recipes[['id', 'name']]
recipe_id_and_rating = recipes_with_reviews[['recipe_id', 'rating']]
merged_table = pd.merge(id_and_name, recipe_id_and_rating, left_on='id', right_on='recipe_id')
print('Объединеная таблица:')
merged_table
```

```python
#Создаю список рецептов, к которым нет отзывов:
recipe_ids = recipes['id']
review_ids = reviews['recipe_id']
recipes_without_reviews = recipe_ids[~recipe_ids.isin(review_ids)]
print('Список рецептов, к которым нет отзывов: ')
recipes_without_reviews
```

```python
#Пробуем найти в таблице рецепт, который не имеет отзывов:
print('Есть ли в объединенной таблице рецепт с id = 223349?')
any(merged_table['id'] == 223349)
```

5.2 При помощи объединения таблиц и группировок, создайте `DataFrame`, состоящий из трех столбцов: `recipe_id`, `name`, `review_count`, где столбец `review_count` содержит кол-во отзывов, оставленных на рецепт `recipe_id`. У рецептов, на которые не оставлен ни один отзыв, в столбце `review_count` должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и найдя строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
#Создаю таблицу со столбцами recipe_id и review_count для рецептов, которые имеют отзывы
review_counter = reviews[['recipe_id', 'review']].groupby('recipe_id').count()
review_counter.reset_index(inplace=True)
review_counter = review_counter.rename(columns={'review': 'review_count'})
#Создаю таблицу со столбцами recipe_id и name
id_and_name = recipes[['id', 'name']]
id_and_name = id_and_name.rename(columns={'id': 'recipe_id'})
#Создаю таблицу со столбцами recipe_id, name и review_count для рецептов, которые имеют отзывы
id_name_reviewcount = pd.merge(id_and_name, review_counter)
#Создаю таблицу со столбцами recipe_id и review_count для рецептов, которые не имеют отзывов
recipes_with_zero_reviews = pd.DataFrame({'recipe_id': recipes_without_reviews, 'review_count': 0})
#Создаю таблицу со столбцами recipe_id, name и review_count для рецептов, которые не имеют отзывов
id_name_reviewcount_zero = pd.merge(id_and_name, recipes_with_zero_reviews)
#Создаю итоговую таблицу
id_name_reviewcount_final = pd.concat([id_name_reviewcount, id_name_reviewcount_zero])
id_name_reviewcount_final
```

```python
#Проверка правильности работы кода
id_name_reviewcount_final[id_name_reviewcount_final['recipe_id'] == 363362]
```

5.3. Выясните, рецепты, добавленные в каком году, имеют наименьший средний рейтинг?

```python
#Создаю таблицу со столбцами id и year
id_and_year = recipes[['id', 'year']].rename(columns={'id': 'recipe_id'})
#Создаю таблицу со столбцами id и rating
id_and_rating = reviews[['recipe_id', 'rating']]
#Создаю таблицу со столбцами id, year и rating
id_year_rating = pd.merge(id_and_year, id_and_rating)
print('Год, в котором были опубликованы рецепты с наименьшим средним рейтингом: ')
id_year_rating.groupby('year')['rating'].mean().idxmin()
```
